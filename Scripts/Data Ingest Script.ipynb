### CREATE INFO DATAFRAME

```python
# loop through the directories inside the data directory
# for each directory:
# open the gnr file
# read the entire data line by line
# for every line:
# check if the data has = sign i.e, ignore userfile line
# rather use the userfile line to append the patient's id with the else tag
# store the data in participants_list for further processing

participants_list = []

for id in os.listdir(data_directory):
    with open(f"{data_directory}/{id}/userfile.gnr", "r") as gnr_file:
        file_data = gnr_file.readlines()
        for data in file_data:
            if '=' in data:
                data = str(data).split('=')
                participants_list.append(data)
            else:
                participants_list.append(['ParticipantID', id])


participants_list
```

```python
titles = []

for i, info in enumerate(participants_list):
    titles.append(info[0])
titles = list(set(titles))
titles
```

```python
# Loop through the participants_list
# take the first element of each element in the list == title for the second element
# store this first element in the titles list
# remove duplicates by using list-set-list conversion




# For every title, create a new key with it
# compare it to the first element of participants_list's current iteration
# if the same, append it to the value of the newly created key and remove new line character if present

participants_dict = {}

for title in titles:
    participants_dict[title] = []
    for info in participants_list:
        if info[0] == title:
            value = info[1]
            if value.endswith("\n"):
                participants_dict[title].append(value.replace('\n', ''))
            else:
                participants_dict[title].append(value)
participants_dict
```

```python
# convert the participants_dict to df

info_df = pd.DataFrame.from_dict(participants_dict)

info_df
```

```python
# create a dictionary
# the keys are: the list of the 24 electrode titles, participant ID

eeg_dict = {}
electrodes = []

# extract the names of the electrodes from signal_headers
for header in signal_headers:
    electrode = header['label']
    electrodes.append(electrode)
len(electrodes)

# make the electrode names keys to the eeg_dictionary
for electrode in electrodes:
    eeg_dict[electrode] = []

eeg_dict
```

```python
# Add participantID to the keys

eeg_dict['ParticipantID'] = []
eeg_dict
```

```python
# fill the dict with the first signal in signals
for index, signal in enumerate(signals):
    eeg_dict[electrodes[index]] = []
    eeg_dict[electrodes[index]].append(np.median(signal))

# testing
```

```python
# clean up the dictionary
for key in eeg_dict:
        eeg_dict[key] = []

for root, dirs, files in os.walk(data_directory):
        for file in files:
            if file.endswith('.edf'):  # Assuming EEG data files have .edf extension
                file_path = os.path.join(root, file)
                # read the edf file using pyedflib.highlevel
                signals, signal_headers, header = highlevel.read_edf(file_path)
                for index, signal in enumerate(signals):
                    eeg_dict[signal_headers[index]['label']].append(np.median(signal))
                participant_id = file_path.split('\\', 2)[1]
                eeg_dict['ParticipantID'].append(participant_id)


eeg_dict
```
