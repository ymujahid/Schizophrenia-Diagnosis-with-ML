# LOADING OF DATASET IN REQUIRED FORMAT

## CREATE EEG DATAFRAME

##### EXPLORE THE PYEDFLIB'S HIGHLEVEL OUTPUTS

```python
eeg_dict = {'Fp1[1]': [],
 'Fp2[2]': [],
 'F3[3]': [],
 'F4[4]': [],
 'C3[5]': [],
 'C4[6]': [],
 'P3[7]': [],
 'P4[8]': [],
 'O1[9]': [],
 'O2[10]': [],
 'F7[11]': [],
 'F8[12]': [],
 'T3[13]': [],
 'T4[14]': [],
 'T5[15]': [],
 'T6[16]': [],
 'Fz[17]': [],
 'Pz[18]': [],
 'Cz[19]': [],
 'Pg1[20]': [],
 'Pg2[21]': [],
 'EOGR[22]': [],
 'EOOGL[23]': [],
 'EMG[24]': []}

for root, dirs, files in os.walk(data_directory):
        for file in files:
            if file.endswith('.edf'):  # Assuming EEG data files have .edf extension
                file_path = os.path.join(root, file)
                # read the edf file using pyedflib.highlevel
                signals, signal_headers, header = highlevel.read_edf(file_path)
                for index, signal in enumerate(signals):
                    # standardize signal
                    mean = np.mean(signal)
                    std_dev = np.std(signal)
                    signal = (signal - mean) / std_dev
                    eeg_dict[signal_headers[index]['label']].append(np.mean(signal))
                # participant_id = file_path.split('\\', 2)[0]
                # eeg_dict['ParticipantID'].append(participant_id)


eeg_dict
```

```python
eeg_df = pd.DataFrame.from_dict(eeg_dict)
```
